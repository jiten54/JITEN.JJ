```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima

# Sample dataset (Replace this with your actual data)
data = {
    "week": pd.date_range(start="2024-01-01", periods=20, freq="W"),
    "units_sold": np.random.randint(100, 300, size=20)
}
data = pd.DataFrame(data)

# Convert 'week' column to datetime format & set as index
data["week"] = pd.to_datetime(data["week"])
data.set_index("week", inplace=True)

# Ensure data is in weekly frequency
data = data.asfreq('W')
data.fillna(method='ffill', inplace=True)

# Data Exploration
print("Data Head:")
print(data.head())

# Plot Data
plt.figure(figsize=(12, 5))
plt.plot(data, label='Product Demand')
plt.xlabel("Date")
plt.ylabel("Demand")
plt.title("Demand Over Time")
plt.legend()
plt.show()

# Train-Test Split
train_size = int(len(data) * 0.8)
train, test = data[:train_size], data[train_size:]

# Auto ARIMA Model Selection
model_auto = auto_arima(train, seasonal=True, stepwise=True, suppress_warnings=True)
print("Best ARIMA Model:", model_auto)

# Fit ARIMA Model
model = ARIMA(train, order=model_auto.order)
model_fit = model.fit()
print(model_fit.summary())

# Forecasting
forecast = model_fit.forecast(steps=len(test))

# Plot Predictions
plt.figure(figsize=(12, 5))
plt.plot(train, label="Train")
plt.plot(test, label="Actual")
plt.plot(test.index, forecast, label="Forecast", linestyle="dashed")
plt.xlabel("Date")
plt.ylabel("Demand")
plt.title("Demand Forecasting using ARIMA")
plt.legend()
plt.show()

print("Forecasting completed successfully!")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from prophet import Prophet  # Updated Prophet import

# Sample dataset (Replace this with your actual data)
data = {
    "week": pd.date_range(start="2024-01-01", periods=50, freq="W"),
    "units_sold": np.random.randint(100, 300, size=50)
}
data = pd.DataFrame(data)

# Convert 'week' column to datetime format & set as index
data["week"] = pd.to_datetime(data["week"])
data.set_index("week", inplace=True)

# Feature Engineering: Add external regressors
np.random.seed(42)
data["holiday_effect"] = np.random.randint(0, 2, size=len(data))  # Simulated holiday impact

data = data.asfreq('W')
data.fillna(method='ffill', inplace=True)

# Exploratory Data Analysis (EDA)
print("Data Head:")
print(data.head())
print("\nData Summary:")
print(data.describe())
print("\nMissing Values:")
print(data.isnull().sum())

# Data Visualization
plt.figure(figsize=(12, 5))
sns.lineplot(data=data, x=data.index, y="units_sold", marker="o", label="Units Sold")
sns.scatterplot(data=data, x=data.index, y=data["holiday_effect"] * 50, color="red", label="Holiday Effect")
plt.xlabel("Date")
plt.ylabel("Units Sold")
plt.title("Demand Trend Over Time")
plt.legend()
plt.grid(True)
plt.show()

# Boxplot for Outliers
plt.figure(figsize=(8, 5))
sns.boxplot(y=data["units_sold"], color="lightblue")
plt.title("Distribution of Units Sold")
plt.show()

# Train-Test Split
train_size = int(len(data) * 0.8)
train, test = data[:train_size], data[train_size:]

# Auto ARIMA Model Selection
model_auto = auto_arima(train["units_sold"], seasonal=True, stepwise=True, suppress_warnings=True)
print("Best ARIMA Model:", model_auto)

# Fit ARIMA Model
arima_model = ARIMA(train["units_sold"], order=model_auto.order)
arima_fit = arima_model.fit()
forecast_arima = arima_fit.forecast(steps=len(test))

# Fit SARIMA Model
sarima_model = SARIMAX(train["units_sold"], order=model_auto.order, seasonal_order=(1,1,1,12))
sarima_fit = sarima_model.fit()
forecast_sarima = sarima_fit.forecast(steps=len(test))

# Fit LSTM Model
X_train, y_train = np.array(train.index.factorize()[0]), np.array(train.units_sold)
X_train = X_train.reshape((X_train.shape[0], 1, 1))

lstm_model = Sequential([
    LSTM(50, activation='relu', input_shape=(1, 1)),
    Dense(1)
])
lstm_model.compile(optimizer='adam', loss='mse')
lstm_model.fit(X_train, y_train, epochs=20, verbose=0)

X_test = np.array(test.index.factorize()[0]).reshape(-1, 1, 1)
lstm_forecast = lstm_model.predict(X_test).flatten()

print("Forecasting completed successfully!")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset (Replace this with actual data)
data = {
    "week": pd.date_range(start="2024-01-01", periods=50, freq="W"),
    "units_sold": np.random.randint(100, 300, size=50)
}
data = pd.DataFrame(data)

# Convert 'week' column to datetime format & set as index
data["week"] = pd.to_datetime(data["week"])
data.set_index("week", inplace=True)

data = data.asfreq('W')
data.fillna(method='ffill', inplace=True)

# Visualization 1: Demand Trend Over Time
plt.figure(figsize=(12, 5))
sns.lineplot(data=data, x=data.index, y="units_sold", marker="o", label="Units Sold", color="blue")
plt.xlabel("Date")
plt.ylabel("Units Sold")
plt.title("Demand Trend Over Time")
plt.legend()
plt.grid(True)
plt.show()

# Visualization 2: Moving Average for Smoother Trends
data["moving_avg"] = data["units_sold"].rolling(window=4).mean()
plt.figure(figsize=(12, 5))
plt.plot(data.index, data["units_sold"], label="Actual Demand", linestyle="dotted", marker="o")
plt.plot(data.index, data["moving_avg"], label="4-Week Moving Average", linestyle="solid", color="red")
plt.xlabel("Date")
plt.ylabel("Units Sold")
plt.title("Demand Forecasting with Moving Average")
plt.legend()
plt.grid(True)
plt.show()

# Visualization 3: Boxplot for Outliers
plt.figure(figsize=(8, 5))
sns.boxplot(y=data["units_sold"], color="lightblue")
plt.title("Distribution of Units Sold (Outlier Detection)")
plt.show()

# Visualization 4: Histogram of Demand Distribution
plt.figure(figsize=(8, 5))
sns.histplot(data["units_sold"], bins=10, kde=True, color="green")
plt.title("Demand Distribution Histogram")
plt.xlabel("Units Sold")
plt.ylabel("Frequency")
plt.show()

print("Visualizations for Demand Forecasting Generated Successfully!")
```
