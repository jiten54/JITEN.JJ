🌍 Global Snakebite Statistics
81,000 to 138,000 deaths occur worldwide each year due to snakebites.
Up to 400,000 survivors suffer from permanent disabilities such as amputations, blindness, and chronic pain.
South Asia, Africa, and Latin America have the highest snakebite cases.
🇮🇳 India Snakebite Data
India accounts for nearly 50% of global snakebite deaths.
1.2 million deaths occurred in India from 2000 to 2019 due to snakebites (about 58,000 deaths per year).
Most cases occur in rural areas during farming activities.
Key venomous snakes in India (Big Four):
Indian Cobra (Naja naja)
Russell’s Viper (Daboia russelii)
Saw-scaled Viper (Echis carinatus)
Common Krait (Bungarus caeruleus)
🐍 Causes of High Snakebite Fatalities
Delayed medical treatment: Many victims rely on traditional healers before seeking hospital care.
Lack of anti-venom: Some rural hospitals have limited access to snakebite treatment.
Poor identification of snake species: People don’t know whether a snake is venomous or not.
🛠️ AI & Snakebite Prevention
AI-powered snake identification apps can help differentiate venomous and non-venomous snakes using images.
Faster identification = quicker treatment = more lives saved.

Step 1: Import Required Libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

Step 2: Create Sample Dataset

```python
# Create a synthetic dataset with species, length, weight, venom status, and region
data = {
    'Species': np.random.choice(['Cobra', 'Viper', 'Python', 'Krait', 'Boa'], 50),
    'Length (cm)': np.random.randint(50, 300, 50),  # Random length values in cm
    'Weight (kg)': np.random.uniform(0.5, 15, 50),  # Random weight values in kg
    'Venomous': np.random.choice([0, 1], 50),  # 0 = Non-venomous, 1 = Venomous
    'Region': np.random.choice(['South India', 'North India', 'East India', 'West India'], 50),
    'Sighting Frequency': np.random.randint(1, 100, 50)  # Random sightings count
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Introduce some missing values randomly
df.loc[np.random.randint(0, 50, 5), 'Weight (kg)'] = np.nan  # Introducing NaNs in Weight
df.loc[np.random.randint(0, 50, 3), 'Region'] = np.nan  # Introducing NaNs in Region

# Display first 5 rows
print(df.head())
```

Step 3: Handle Missing Values

```python
# Fill missing values in 'Weight (kg)' with the mean
df['Weight (kg)'].fillna(df['Weight (kg)'].mean(), inplace=True)

# Fill missing values in 'Region' with the mode (most frequent region)
df['Region'].fillna(df['Region'].mode()[0], inplace=True)

# Check if any missing values remain
print(df.isnull().sum())
```

Step 4: Encode Categorical Variables

```python
# Encode the 'Species' column
le_species = LabelEncoder()
df['Species'] = le_species.fit_transform(df['Species'])  # Convert species to numbers

# Encode the 'Region' column
le_region = LabelEncoder()
df['Region'] = le_region.fit_transform(df['Region'])  # Convert regions to numbers

# Display encoded values
print(df.head())
```

Step 5: Normalize Numerical Features

```python
# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Apply scaling on numerical columns
df[['Length (cm)', 'Weight (kg)']] = scaler.fit_transform(df[['Length (cm)', 'Weight (kg)']])

# Display scaled dataset
print(df.head())
```

 Step 2: Outlier Detection & Handling

```python
import numpy as np

# Define function to remove outliers using IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# Apply outlier removal to numerical columns
for col in ['Length (cm)', 'Weight (kg)', 'Sighting Frequency']:
    df = remove_outliers(df, col)

# Boxplot to visualize data after outlier removal
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
sns.boxplot(data=df[['Length (cm)', 'Weight (kg)', 'Sighting Frequency']], palette='coolwarm')
plt.title("Boxplot After Outlier Removal")
plt.show()
```

 Step 3: Feature Engineering
I create new meaningful features to improve model accuracy.

🔹 Feature 1: Body Mass Index for Snakes

```python
df['Weight-to-Length Ratio'] = df['Weight (kg)'] / df['Length (cm)']
```

 Feature 2: Categorizing Snakes Based on Size

```python
def categorize_size(length):
    if length < 0.3:
        return 'Small'
    elif length < 0.7:
        return 'Medium'
    else:
        return 'Large'

df['Size Category'] = df['Length (cm)'].apply(categorize_size)
```

 Feature 3: Sightings as High/Low Frequency

```python
df['High Sighting Frequency'] = df['Sighting Frequency'].apply(lambda x: 1 if x > df['Sighting Frequency'].median() else 0)
```

Step 4: Hypothesis Testing

```python
from scipy.stats import ttest_ind

venomous = df[df['Venomous'] == 1]['Weight (kg)']
non_venomous = df[df['Venomous'] == 0]['Weight (kg)']

t_stat, p_val = ttest_ind(venomous, non_venomous)

print(f"T-Test for Weight of Venomous vs. Non-Venomous Snakes:\nT-Statistic = {t_stat}, P-Value = {p_val}")
```

Hypothesis 2: Are Some Snake Species More Likely to Be Venomous?

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['Species'], df['Venomous'])
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Square Test for Species vs. Venomous:\nChi2-Statistic = {chi2_stat}, P-Value = {p_val}")
```
